```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import itertools
```

Question: Are normal distributions circularly symmetric when I apply them separately to the two axes?

```python
n = 1000
points = np.random.normal(size=(2,n))
rt = np.sqrt(2)/2
points2= np.dot([[rt,-rt],[rt,rt]],np.random.normal(size=(2,n)))


plt.scatter(*points);
plt.scatter(*points2,c='red');
plt.axes().set_aspect('equal');
```

Answer: as much as I care, yes
